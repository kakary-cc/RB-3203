```python
import cv2
import common
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt # this lets you draw inline pictures in the notebooks
import pylab # this allows you to control figure size
pylab.rcParams['figure.figsize'] = (15.0, 12.0) # this controls figure size in the notebook
```

```python
base_image = cv2.imread('images/NewYork427.png')
base_image = cv2.resize(base_image, (1080, 720))
grey = cv2.cvtColor(base_image, cv2.COLOR_BGR2GRAY)
```

```python
test_image = cv2.imread('images/NewYork427.png')
test_image = cv2.resize(test_image, (1080, 720))
body_cascade = cv2.CascadeClassifier('haarcascade_fullbody.xml')
# detectMultiScale(image, scalefactor, minNeighbors)
bodies = body_cascade.detectMultiScale(grey, 1.1, 2)
for (x,y,w,h) in bodies:
    cv2.rectangle(test_image,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB))
```

```python
# test_image = cv2.imread('images/NewYork427.png')
# test_image = cv2.resize(test_image, (1080, 720))
face_cascade = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
# detectMultiScale(image, scalefactor, minNeighbors)
faces = face_cascade.detectMultiScale(grey, 1.1, 1)
for (x,y,w,h) in faces:
    cv2.rectangle(test_image,(x,y),(x+w,y+h),(0,255,0),2)
plt.imshow(cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB))
```

```python
# test_image = cv2.imread('images/NewYork427.png')
# test_image = cv2.resize(test_image, (1080, 720))
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
# detectMultiScale(image, scalefactor, minNeighbors)
eyes = eye_cascade.detectMultiScale(grey, 1.6, 2)
for (x,y,w,h) in eyes:
    cv2.rectangle(test_image,(x,y),(x+w,y+h),(0,0,255),2)
plt.imshow(cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB))
```
